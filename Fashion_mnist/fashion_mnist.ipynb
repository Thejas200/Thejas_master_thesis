```python


import os
import math
import time
import json
import random
import shutil
from dataclasses import dataclass, field
from typing import Dict, List, Tuple, Optional

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display

# Reduce TF/absl noise
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# ------------------------------- Reproducibility -------------------------------
GLOBAL_SEED = 42
random.seed(GLOBAL_SEED)
np.random.seed(GLOBAL_SEED)
tf.random.set_seed(GLOBAL_SEED)

# Make GPU memory growth (prevents huge upfront alloc)
try:
    gpus = tf.config.list_physical_devices("GPU")
    if gpus:
        for g in gpus:
            tf.config.experimental.set_memory_growth(g, True)
except Exception:
    pass

# ------------------------------- IO / Paths -----------------------------------
ROOT_OUT_DIR = "ga_markov_outputs"
RUN_ID = time.strftime("%Y%m%d_%H%M%S")
OUT_DIR = os.path.join(ROOT_OUT_DIR, f"run_{RUN_ID}")
FIG_DIR = os.path.join(OUT_DIR, "figs")
os.makedirs(FIG_DIR, exist_ok=True)
CLASS_NAMES = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot",
]

# ------------------------------- Dataset --------------------------------------
def load_data():
    (x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
    x_train = x_train.astype("float32") / 255.0
    x_test  = x_test.astype("float32") / 255.0
    x_train = np.expand_dims(x_train, -1)
    x_test  = np.expand_dims(x_test, -1)
    # Train/Val split
    val_frac = 0.1
    n = x_train.shape[0]
    idx = np.arange(n)
    np.random.shuffle(idx)
    split = int(n * (1 - val_frac))
    tr_idx, val_idx = idx[:split], idx[split:]
    x_tr, y_tr = x_train[tr_idx], y_train[tr_idx]
    x_val, y_val = x_train[val_idx], y_train[val_idx]
    return (x_tr, y_tr), (x_val, y_val), (x_test, y_test)

def summarize_data_splits(x_tr, y_tr, x_val, y_val, x_te, y_te, fig_path: str):
    split_names = ["Train", "Validation", "Test"]
    split_counts = [int(len(y_tr)), int(len(y_val)), int(len(y_te))]
    total = int(sum(split_counts))
    split_percents = [100.0 * c / total for c in split_counts]

    summary_df = pd.DataFrame({
        "Split": split_names,
        "Samples": split_counts,
        "Percent": [round(v, 2) for v in split_percents],
        "Image shape": [str(x_tr.shape[1:]), str(x_val.shape[1:]), str(x_te.shape[1:])],
    })
    print("\nDataset split summary:")
    print(summary_df.to_string(index=False))

    class_counts = {
        "Train": np.bincount(y_tr, minlength=len(CLASS_NAMES)),
        "Validation": np.bincount(y_val, minlength=len(CLASS_NAMES)),
        "Test": np.bincount(y_te, minlength=len(CLASS_NAMES)),
    }
    class_df = pd.DataFrame({
        "Class": CLASS_NAMES,
        "Train": class_counts["Train"],
        "Validation": class_counts["Validation"],
        "Test": class_counts["Test"],
    })
    print("\nClass distribution across splits:")
    print(class_df.to_string(index=False))

    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    colors = ["tab:blue", "tab:orange", "tab:green"]
    bars = axes[0].bar(split_names, split_counts, color=colors, alpha=0.9)
    axes[0].set_title("Number of Images in Each Split")
    axes[0].set_ylabel("Number of samples")
    axes[0].grid(True, axis="y", alpha=0.3)
    for bar, count, pct in zip(bars, split_counts, split_percents):
        axes[0].text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + max(split_counts) * 0.015,
            f"{count:,}\n({pct:.1f}%)",
            ha="center",
            va="bottom",
            fontsize=10,
        )

    x = np.arange(len(CLASS_NAMES))
    width = 0.26
    axes[1].bar(x - width, class_counts["Train"], width=width, label="Train", color="tab:blue")
    axes[1].bar(x, class_counts["Validation"], width=width, label="Validation", color="tab:orange")
    axes[1].bar(x + width, class_counts["Test"], width=width, label="Test", color="tab:green")
    axes[1].set_title("Class Distribution by Split")
    axes[1].set_ylabel("Number of samples")
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(CLASS_NAMES, rotation=35, ha="right")
    axes[1].grid(True, axis="y", alpha=0.3)
    axes[1].legend()

    fig.suptitle("Fashion-MNIST Train / Validation / Test Overview", fontsize=15)
    fig.tight_layout(rect=[0, 0.02, 1, 0.95])
    fig.savefig(fig_path, dpi=180, bbox_inches="tight")
    plt.show()
    plt.close(fig)

    return {
        "splits": {
            name.lower(): {
                "samples": count,
                "percent": round(pct, 2),
            }
            for name, count, pct in zip(split_names, split_counts, split_percents)
        },
        "figure_path": fig_path,
    }

def save_learning_curves_figure(history, test_acc: float, test_loss: float, fig_path: str, title_prefix: str = "Baseline"):
    epochs = np.arange(1, len(history.history["accuracy"]) + 1)
    fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))

    axes[0].plot(epochs, history.history["accuracy"], marker="o", linewidth=2, label="Train accuracy")
    axes[0].plot(epochs, history.history["val_accuracy"], marker="s", linewidth=2, label="Validation accuracy")
    axes[0].axhline(test_acc, color="tab:green", linestyle="--", linewidth=2, label=f"Test accuracy = {test_acc:.4f}")
    axes[0].set_title(f"Accuracy During {title_prefix} Training")
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("Accuracy")
    axes[0].set_xticks(epochs)
    axes[0].grid(True, alpha=0.3)
    axes[0].legend()

    axes[1].plot(epochs, history.history["loss"], marker="o", linewidth=2, label="Train loss")
    axes[1].plot(epochs, history.history["val_loss"], marker="s", linewidth=2, label="Validation loss")
    axes[1].axhline(test_loss, color="tab:red", linestyle="--", linewidth=2, label=f"Test loss = {test_loss:.4f}")
    axes[1].set_title(f"Loss During {title_prefix} Training")
    axes[1].set_xlabel("Epoch")
    axes[1].set_ylabel("Loss")
    axes[1].set_xticks(epochs)
    axes[1].grid(True, alpha=0.3)
    axes[1].legend()

    fig.suptitle(f"{title_prefix}: Train, Validation, and Test Performance", fontsize=15)
    fig.tight_layout(rect=[0, 0.02, 1, 0.95])
    fig.savefig(fig_path, dpi=180, bbox_inches="tight")
    plt.show()
    plt.close(fig)

def display_saved_figure(fig_path: str, heading: str):
    print(f"\n{heading}: {fig_path}")
    if os.path.exists(fig_path):
        display(Image(filename=fig_path))

def confusion_matrix_np(y_true: np.ndarray, y_pred: np.ndarray, num_classes: int = 10) -> np.ndarray:
    cm = np.zeros((num_classes, num_classes), dtype=int)
    for yt, yp in zip(y_true, y_pred):
        cm[int(yt), int(yp)] += 1
    return cm

def per_class_accuracy(cm: np.ndarray) -> np.ndarray:
    totals = cm.sum(axis=1)
    return np.divide(np.diag(cm), totals, out=np.zeros_like(totals, dtype=float), where=totals > 0)

def save_final_comparison_figure(x_te, y_te, baseline_pred, final_pred, fig_path: str):
    cm_base = confusion_matrix_np(y_te, baseline_pred, num_classes=len(CLASS_NAMES))
    cm_final = confusion_matrix_np(y_te, final_pred, num_classes=len(CLASS_NAMES))
    base_pc = per_class_accuracy(cm_base)
    final_pc = per_class_accuracy(cm_final)
    delta_pc = final_pc - base_pc

    improved = np.where((baseline_pred != y_te) & (final_pred == y_te))[0]
    difficult = np.where(final_pred != y_te)[0]
    chosen = improved[:4]
    if len(chosen) < 4:
        chosen = np.concatenate([chosen, difficult[: 4 - len(chosen)]])
    if len(chosen) < 4:
        chosen = np.concatenate([chosen, np.arange(min(len(y_te), 4 - len(chosen)))])
    chosen = chosen[:4]

    fig = plt.figure(figsize=(18, 10))
    gs = fig.add_gridspec(2, 3, height_ratios=[3.2, 2.2], width_ratios=[1.15, 1.15, 1.0])

    ax0 = fig.add_subplot(gs[0, 0])
    im0 = ax0.imshow(cm_base, cmap="Blues")
    ax0.set_title("Baseline Confusion Matrix")
    ax0.set_xlabel("Predicted")
    ax0.set_ylabel("True")
    ax0.set_xticks(range(len(CLASS_NAMES)))
    ax0.set_yticks(range(len(CLASS_NAMES)))
    ax0.set_xticklabels(range(len(CLASS_NAMES)))
    ax0.set_yticklabels(CLASS_NAMES, fontsize=9)
    fig.colorbar(im0, ax=ax0, fraction=0.046, pad=0.04)

    ax1 = fig.add_subplot(gs[0, 1])
    im1 = ax1.imshow(cm_final, cmap="Greens")
    ax1.set_title("Final Confusion Matrix")
    ax1.set_xlabel("Predicted")
    ax1.set_ylabel("True")
    ax1.set_xticks(range(len(CLASS_NAMES)))
    ax1.set_yticks(range(len(CLASS_NAMES)))
    ax1.set_xticklabels(range(len(CLASS_NAMES)))
    ax1.set_yticklabels(CLASS_NAMES, fontsize=9)
    fig.colorbar(im1, ax=ax1, fraction=0.046, pad=0.04)

    ax2 = fig.add_subplot(gs[0, 2])
    ypos = np.arange(len(CLASS_NAMES))
    bar_h = 0.36
    ax2.barh(ypos - bar_h / 2, base_pc * 100.0, height=bar_h, color="tab:blue", label="Baseline")
    ax2.barh(ypos + bar_h / 2, final_pc * 100.0, height=bar_h, color="tab:green", label="Final")
    ax2.set_yticks(ypos)
    ax2.set_yticklabels(CLASS_NAMES, fontsize=9)
    ax2.set_xlabel("Per-class accuracy (%)")
    ax2.set_title("Per-Class Accuracy Comparison")
    ax2.set_xlim(0, 100)
    ax2.grid(True, axis="x", alpha=0.3)
    ax2.legend(loc="lower right", fontsize=8)
    for i, v in enumerate(delta_pc):
        delta_label = f"{v * 100.0:+.1f}"
        text_x = min(max(base_pc[i], final_pc[i]) * 100.0 + 1.2, 98.0)
        ax2.text(text_x, ypos[i], delta_label, va="center", fontsize=8, color="tab:green" if v >= 0 else "tab:red")

    sample_gs = gs[1, :].subgridspec(1, 4)
    for i, idx in enumerate(chosen):
        ax = fig.add_subplot(sample_gs[0, i])
        ax.imshow(x_te[int(idx)].squeeze(), cmap="gray")
        ax.set_title(
            f"True: {CLASS_NAMES[int(y_te[int(idx)])]}\n"
            f"Base: {CLASS_NAMES[int(baseline_pred[int(idx)])]}\n"
            f"Final: {CLASS_NAMES[int(final_pred[int(idx)])]}",
            fontsize=10,
        )
        ax.axis("off")

    fig.suptitle("Fashion-MNIST Baseline vs Final Model", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.savefig(fig_path, dpi=150)
    plt.close(fig)

# ------------------------------- Models ---------------------------------------
def build_baseline(input_shape=(28, 28, 1), num_classes=10):
    model = keras.Sequential(
        [
            keras.Input(shape=input_shape),
            layers.Conv2D(32, 3, activation="relu"),
            layers.MaxPooling2D(),
            layers.Conv2D(64, 3, activation="relu"),
            layers.MaxPooling2D(),
            layers.Flatten(),
            layers.Dense(128, activation="relu"),
            layers.Dropout(0.25),
            layers.Dense(num_classes, activation="softmax"),
        ]
    )
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model

def build_model_from_params(p: Dict, input_shape=(28,28,1), num_classes=10):
    # p keys: filters, num_conv_blocks, kernel_size, pooling_type, dense_units, dropout, use_batchnorm, batch_size
    model = keras.Sequential()
    model.add(keras.Input(shape=input_shape))
    filters = int(p["filters"])
    k = p.get("kernel_size", (3,3))
    for _ in range(int(p["num_conv_blocks"])):
        model.add(layers.Conv2D(filters, k, padding="same", activation="relu"))
        if p.get("use_batchnorm", False):
            model.add(layers.BatchNormalization())
        model.add(layers.Conv2D(filters, k, padding="same", activation="relu"))
        if p.get("use_batchnorm", False):
            model.add(layers.BatchNormalization())
        if p["pooling_type"] == "max":
            model.add(layers.MaxPooling2D(pool_size=(2,2)))
        else:
            model.add(layers.AveragePooling2D(pool_size=(2,2)))
        filters = max(filters // 2, 16)  # gentle taper

    model.add(layers.Flatten())
    model.add(layers.Dense(int(p["dense_units"]), activation="relu"))
    model.add(layers.Dropout(float(p["dropout"])))
    model.add(layers.Dense(num_classes, activation="softmax"))
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model

def train_and_eval(model, x_tr, y_tr, x_val, y_val, epochs=6, batch_size=64, verbose=0):
    early_stop = keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True,
        min_delta=1e-4,
    )
    history = model.fit(
        x_tr, y_tr,
        validation_data=(x_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=verbose,
        callbacks=[early_stop],
    )
    val_acc = float(np.max(history.history["val_accuracy"]))
    return val_acc

# ------------------------------- GA Setup -------------------------------------
@dataclass
class Individual:
    params: Dict
    fitness: Optional[float] = None
    id: int = field(default=-1)

def sample_params() -> Dict:
    return {
        "filters": random.choice([32, 48, 64, 96]),
        "num_conv_blocks": random.choice([1, 2]),
        "kernel_size": (3,3),
        "pooling_type": random.choice(["max", "avg"]),
        "dense_units": random.choice([64, 128, 192, 256, 384]),
        "dropout": round(float(np.clip(np.random.normal(0.25, 0.1), 0.05, 0.45)), 3),
        "batch_size": random.choice([64, 128]),
        "use_batchnorm": random.choice([0, 1]) == 1,
    }

def mutate_params(p: Dict) -> Dict:
    q = dict(p)
    if random.random() < 0.4:
        q["filters"] = random.choice([32, 48, 64, 96])
    if random.random() < 0.4:
        q["num_conv_blocks"] = random.choice([1, 2])
    if random.random() < 0.4:
        q["pooling_type"] = random.choice(["max", "avg"])
    if random.random() < 0.5:
        q["dense_units"] = random.choice([64, 128, 192, 256, 384])
    if random.random() < 0.5:
        q["dropout"] = round(float(np.clip(q["dropout"] + np.random.normal(0, 0.05), 0.05, 0.5)), 3)
    if random.random() < 0.3:
        q["batch_size"] = random.choice([64, 128])
    if random.random() < 0.3:
        q["use_batchnorm"] = not q["use_batchnorm"]
    return q

def crossover_params(p1: Dict, p2: Dict) -> Dict:
    child = {}
    for k in p1.keys():
        child[k] = p1[k] if random.random() < 0.5 else p2[k]
    return child

def rank_bins(ranks: List[int], num_bins: int = 10) -> np.ndarray:
    """Map rank (0 best … n-1 worst) to bin index 0..num_bins-1."""
    n = len(ranks)
    bins = np.zeros(n, dtype=int)
    for i, r in enumerate(ranks):
        bins[i] = min((r * num_bins) // n, num_bins - 1)
    return bins

# --------------------- Markov utilities (robust & interpretable) ---------------------
def _smooth_and_normalize_rows(P: np.ndarray, counts: np.ndarray, alpha: float) -> np.ndarray:
    """
    Dirichlet/Laplace smoothing per row. For empty rows, fallback to uniform.
    Returns a valid row-stochastic matrix.
    """
    k = P.shape[1]
    Q = P.astype(float).copy()
    for i in range(Q.shape[0]):
        if counts[i] > 0:
            Q[i, :] = (Q[i, :] + alpha) / (counts[i] + alpha * k)
        else:
            Q[i, :] = 1.0 / k
    return Q

def _lazy_matrix(P: np.ndarray, lazy: bool) -> np.ndarray:
    if not lazy:
        return P
    I = np.eye(P.shape[0], dtype=float)
    return 0.5 * (I + P)

def compute_markov_and_lambda2(
    parent_bins: List[int],
    child_bins: List[int],
    num_bins: int = 10,
    alpha: float = 1e-2,
    use_lazy: bool = False, # keep False 
) -> Tuple[np.ndarray, float, float, np.ndarray]:
    """
    Build smoothed (and optional lazy) transition matrix P, return:
      - P (row-stochastic), lambda2 (SLEM), relaxation_time, counts per parent bin.
    """
    P_raw = np.zeros((num_bins, num_bins), dtype=float)
    counts = np.zeros(num_bins, dtype=float)
    for pb, cb in zip(parent_bins, child_bins):
        P_raw[pb, cb] += 1.0
        counts[pb] += 1.0

    # Smooth + normalize (uniform fallback for empty rows)
    P = _smooth_and_normalize_rows(P_raw, counts, alpha=alpha)

    # Optional lazy chain to stabilize spectrum & avoid periodicity
    P = _lazy_matrix(P, lazy=use_lazy)

    # Eigenvalues: SLEM (second largest eigenvalue modulus)
    w = np.linalg.eigvals(P)
    w = np.sort(np.abs(w))[::-1]
    lambda2 = float(w[1]) if len(w) > 1 else 0.0

    # Relaxation time from the absolute spectral gap
    relaxation_time = float(1.0 / max(1e-9, (1.0 - min(lambda2, 0.999999))))

    return P, lambda2, relaxation_time, counts

def print_matrix(P):
    with np.printoptions(precision=3, suppress=True):
        print(P)

def save_transition_heatmap(P: np.ndarray, gen: int, fig_dir: str):
    plt.figure(figsize=(5,4))
    ax = plt.gca()
    im = ax.imshow(P, aspect="auto", origin="upper")
    plt.title(f"Gen {gen:02d} Markov Transition (parent bin → child bin)")
    plt.xlabel("Child rank bin")
    plt.ylabel("Parent rank bin")
    plt.colorbar(im, fraction=0.046, pad=0.04)
    path = os.path.join(fig_dir, f"gen_{gen:02d}_transition.png")
    plt.tight_layout()
    plt.savefig(path, dpi=150)
    plt.close()

def population_to_df(pop: List[Individual]) -> pd.DataFrame:
    rows = []
    for ind in pop:
        p = ind.params
        rows.append({
            "id": ind.id,
            "fitness": ind.fitness,
            "filters": p["filters"],
            "blocks": p["num_conv_blocks"],
            "kernel": f"{p['kernel_size']}",
            "pool": p["pooling_type"],
            "dense": p["dense_units"],
            "drop": p["dropout"],
            "bs": p["batch_size"],
            "bn": int(p["use_batchnorm"]),
        })
    df = pd.DataFrame(rows).sort_values(by="fitness", ascending=False)
    return df

def print_top_table(pop: List[Individual], k: int = 10):
    df = population_to_df(pop).head(k)
    print("Top individuals this generation:")
    print("rank id    fitness  filters blocks pool dense drop  bs  bn ")
    print("-----------------------------------------------------------")
    for i, row in enumerate(df.itertuples(index=False), 1):
        print(f"{i:<5}{int(row.id):<6}{row.fitness:>7.4f}   {int(row.filters):<7}{int(row.blocks):<6} {row.pool:<4} {int(row.dense):<5} {row.drop:<4} {int(row.bs):<3} {int(row.bn):<2}")

def params_key(p: Dict) -> Tuple:
    return (
        int(p["filters"]),
        int(p["num_conv_blocks"]),
        tuple(p["kernel_size"]),
        str(p["pooling_type"]),
        int(p["dense_units"]),
        float(p["dropout"]),
        int(p["batch_size"]),
        bool(p["use_batchnorm"]),
    )

# ------------------------------- GA Loop --------------------------------------
def evaluate_individual(ind: Individual, x_tr, y_tr, x_val, y_val, epochs=6, verbose=0) -> float:
    m = build_model_from_params(ind.params)
    acc = train_and_eval(
        m, x_tr, y_tr, x_val, y_val,
        epochs=epochs, batch_size=ind.params["batch_size"], verbose=verbose
    )
    ind.fitness = acc
    return acc

def run_ga(
    x_tr, y_tr, x_val, y_val,
    pop_size=40, max_gens=30, epochs_per_eval=6,
    elite_frac=0.15, mutation_prob=0.45, crossover_frac=0.7,
    patience=6, min_delta=0.0008,
    verbose_eval=False,
    # --- Markov robustness knobs ---
    markov_alpha: float = 0.1,   # stronger smoothing
    markov_lazy: bool = False,   # NO lazy matrix
    low_sample_threshold: int = 3,
    num_bins: int = 10,
    finalist_k: int = 3,
):
    rng = random.Random(GLOBAL_SEED)
    next_id = 0

    # Init population
    population: List[Individual] = []
    for _ in range(pop_size):
        ind = Individual(params=sample_params(), id=next_id)
        next_id += 1
        population.append(ind)

    best_acc = -1.0
    best_ind: Optional[Individual] = None
    stagnation = 0

    print(f"\nStarting GA with {pop_size} individuals for up to {max_gens} generations\n")

    # Evaluate initial population
    for ind in population:
        if ind.fitness is None:
            evaluate_individual(ind, x_tr, y_tr, x_val, y_val, epochs=epochs_per_eval, verbose=0 if not verbose_eval else 2)

    # GA generations
    for gen in range(1, max_gens + 1):
        print(f"=== Generation {gen} ===\n")

        # Sort by fitness
        population.sort(key=lambda z: z.fitness if z.fitness is not None else -1e9, reverse=True)

        # Parent rank bins from current pop ranking
        parent_ranks = list(range(len(population)))  # 0 best .. n-1 worst
        parent_bins = rank_bins(parent_ranks, num_bins=num_bins)

        # Elitism
        elite_k = max(1, int(elite_frac * pop_size))
        elites = population[:elite_k]

        # --------- Fitness-Proportionate (Roulette-Wheel) Selection ----------
        # We sample parents with probability proportional to fitness.
        # Since fitness is val_accuracy in (0,1), it's already positive, but we still add eps
        # for numerical safety if anything is 0.0.
        def make_roulette_selector(pop: List[Individual]):
            fits = np.array([float(ind.fitness) for ind in pop], dtype=float)
            eps = 1e-12
            fits = np.maximum(fits, 0.0) + eps
            total = float(np.sum(fits))
            if not np.isfinite(total) or total <= 0.0:
                # fallback: uniform selection
                cdf = np.cumsum(np.ones(len(pop)) / len(pop))
            else:
                probs = fits / total
                cdf = np.cumsum(probs)
                cdf[-1] = 1.0  # guard against rounding drift
            def select_one():
                r = rng.random()
                idx = int(np.searchsorted(cdf, r, side="right"))
                if idx >= len(pop):
                    idx = len(pop) - 1
                return pop[idx]
            return select_one

        roulette_select = make_roulette_selector(population)
        # ------------------------------------------------------------------------------

        # Create next generation
        children: List[Individual] = []
        # Markov samples: (child_id, parent_bin). For crossover, record BOTH parents.
        created_child_records: List[Tuple[int, int]] = []

        # Carry elites (clone)
        for e in elites:
            child = Individual(params=dict(e.params), id=next_id)
            next_id += 1
            children.append(child)
            created_child_records.append((child.id, int(parent_bins[population.index(e)])))

        # Fill rest by crossover/mutation
        while len(children) < pop_size:
            if rng.random() < crossover_frac:
                p1 = roulette_select()
                p2 = roulette_select()
                child_params = crossover_params(p1.params, p2.params)
                if rng.random() < mutation_prob:
                    child_params = mutate_params(child_params)

                # both-parent attribution for Markov
                p1_bin = int(parent_bins[population.index(p1)])
                p2_bin = int(parent_bins[population.index(p2)])

                child = Individual(params=child_params, id=next_id)
                next_id += 1
                children.append(child)
                created_child_records.append((child.id, p1_bin))
                created_child_records.append((child.id, p2_bin))

            else:
                p = roulette_select()
                child_params = mutate_params(p.params) if rng.random() < mutation_prob else dict(p.params)
                parent_bin_for_markov = int(parent_bins[population.index(p)])

                child = Individual(params=child_params, id=next_id)
                next_id += 1
                children.append(child)
                created_child_records.append((child.id, parent_bin_for_markov))

        # Evaluate children (quiet for clean gen summaries unless verbose_eval=True)
        for ch in children:
            evaluate_individual(ch, x_tr, y_tr, x_val, y_val, epochs=epochs_per_eval, verbose=0 if not verbose_eval else 2)

        # Sort children by fitness to define child rank-bins
        children.sort(key=lambda z: z.fitness, reverse=True)
        child_ranks_sorted = list(range(len(children)))
        child_bins_sorted = rank_bins(child_ranks_sorted, num_bins=num_bins)

        # Map child id -> its child bin after sorting
        id_to_childbin = {children[i].id: int(child_bins_sorted[i]) for i in range(len(children))}

        # Build aligned sequences: (parent bin, child bin) for ALL recorded samples
        parent_seq = []
        child_seq = []
        for (cid, pbin) in created_child_records:
            parent_seq.append(int(pbin))
            child_seq.append(int(id_to_childbin[cid]))

        # Compute Markov P, lambda2, relaxation time
        P, lambda2, relaxation_time, counts = compute_markov_and_lambda2(
            parent_seq, child_seq, num_bins=num_bins, alpha=markov_alpha, use_lazy=markov_lazy
        )

        # Low-sample warning (pre-smoothing counts)
        low_bins = np.where(counts < low_sample_threshold)[0]
        if len(low_bins) > 0:
            print(f"Note: low-sample parent bins this gen (<{low_sample_threshold} samples): {list(low_bins)}")

        # Log generation stats
        fitnesses = [ind.fitness for ind in children]
        gen_best = float(np.max(fitnesses))
        gen_mean = float(np.mean(fitnesses))

        print("Generation {} Statistics:".format(gen))
        print(f"- Best accuracy:    {gen_best:.4f}")
        print(f"- Average accuracy: {gen_mean:.4f}")
        print(f"- λ₂ (SLEM): {lambda2:.4f}, Relaxation time: {relaxation_time:.2f} generations"
              + (" [lazy]" if markov_lazy else ""))
        print("- Transition matrix (rows=parent rank-bin, cols=child rank-bin):")
        print_matrix(P)

        # Pretty table of top individuals
        print_top_table(children, k=10)

        # Save artifacts
        df = population_to_df(children)
        csv_path = os.path.join(OUT_DIR, f"gen_{gen:02d}_population.csv")
        df.to_csv(csv_path, index=False)
        save_transition_heatmap(P, gen, FIG_DIR)
        print(f"Saved population CSV: {os.path.basename(csv_path)}")
        print(f"Saved transition heatmap: gen_{gen:02d}_transition.png\n")

        # Early stopping tracking
        if gen_best > best_acc + min_delta:
            best_acc = gen_best
            best_ind = max(children, key=lambda z: z.fitness)
            stagnation = 0
        else:
            stagnation += 1

        # Next population
        population = children

        if stagnation >= patience:
            print(f"Early stopping: <{min_delta} improvement for {patience} gens\n")
            break

    population.sort(key=lambda z: z.fitness if z.fitness is not None else -1e9, reverse=True)
    finalists: List[Dict] = [dict(ind.params) for ind in population[:max(1, int(finalist_k))]]

    # Return best and a small finalist pool for a more robust final selection pass.
    return (
        best_ind if best_ind is not None else max(population, key=lambda z: z.fitness),
        best_acc,
        finalists,
    )

# ------------------------------- Main -----------------------------------------
def main():
    # ---------------- Baseline ----------------
    print("Training baseline CNN model (Fashion-MNIST)...")
    (x_tr, y_tr), (x_val, y_val), (x_te, y_te) = load_data()
    split_fig_path = os.path.join(FIG_DIR, "dataset_split_overview.png")
    split_summary = summarize_data_splits(x_tr, y_tr, x_val, y_val, x_te, y_te, split_fig_path)
    baseline = build_baseline()
    # Verbose ON for baseline
    hist = baseline.fit(
        x_tr, y_tr,
        validation_data=(x_val, y_val),
        epochs=17,
        batch_size=64,
        verbose=2,
        callbacks=[
            keras.callbacks.EarlyStopping(
                monitor="val_loss",
                patience=4,
                restore_best_weights=True,
                min_delta=1e-4,
            )
        ],
    )
    val_acc = float(np.max(hist.history["val_accuracy"]))
    baseline_probs = baseline.predict(x_te, verbose=0)
    baseline_pred = np.argmax(baseline_probs, axis=1)
    test_loss, test_acc = baseline.evaluate(x_te, y_te, verbose=0)
    baseline_curves_path = os.path.join(FIG_DIR, "baseline_learning_curves.png")
    save_learning_curves_figure(hist, test_acc, test_loss, baseline_curves_path, title_prefix="Baseline")
    print(f"\nBaseline Validation Accuracy: {val_acc:.4f}")
    print(f"Baseline Test Accuracy: {test_acc:.4f}")

    # ---------------- GA ----------------
    print("\nInitializing Genetic Algorithm for CNN optimization (Fashion-MNIST)...\n")
    best_ind, best_ga_val, _ga_finalists = run_ga(
        x_tr, y_tr, x_val, y_val,
        pop_size=25, max_gens=20, epochs_per_eval=6,
        elite_frac=0.15, mutation_prob=0.45, crossover_frac=0.7,
        patience=6, min_delta=0.0008,
        verbose_eval=False,       # set True to see epoch logs during GA too
        markov_alpha=0.1,         # stronger smoothing (stable λ2)
        markov_lazy=False,        # NO lazy chain 
        low_sample_threshold=3,
        num_bins=10,
        finalist_k=3,
    )

    # ---------------- Final Model Selection ----------------
    winner = {
        "label": "ga_best",
        "params": dict(best_ind.params),
        "val_acc_mean": float(best_ga_val),
        "val_acc_std": 0.0,
        "val_acc_runs": [float(best_ga_val)],
    }
    print("\n=== Final Model Selection (Fashion-MNIST) ===")
    print("Using the single best GA candidate without top-k reselection.")
    print(
        f"Selected GA best by validation accuracy: {winner['label']} "
        f"(val_acc={winner['val_acc_mean']:.4f})"
    )

    print("\n=== Evaluating Final Optimized Model (Fashion-MNIST) ===")
    print("Selected Parameters:")
    for k, v in winner["params"].items():
        print(f"- {k}: {v}")

    best_model = winner.get("model")
    final_history = winner.get("history")
    if best_model is None:
        best_model = build_model_from_params(winner["params"])
        final_history = best_model.fit(
            x_tr, y_tr,
            validation_data=(x_val, y_val),
            epochs=40,
            batch_size=winner["params"]["batch_size"],
            verbose=2,
            callbacks=[
                keras.callbacks.EarlyStopping(
                    monitor="val_loss",
                    patience=5,
                    restore_best_weights=True,
                    min_delta=1e-4,
                )
            ],
        )
    final_probs = best_model.predict(x_te, verbose=0)
    final_pred = np.argmax(final_probs, axis=1)
    final_loss, final_acc = best_model.evaluate(x_te, y_te, verbose=0)
    final_curves_path = os.path.join(FIG_DIR, "final_model_learning_curves.png")
    print(f"USING NEW FINAL CURVE CODE: {final_curves_path}")
    if final_history is not None:
        save_learning_curves_figure(final_history, final_acc, final_loss, final_curves_path, title_prefix="Final optimized model")
    print(f"\nFinal Test Accuracy: {final_acc:.4f}")
    print(f"Improvement over baseline test acc: {(final_acc - test_acc)*100:.2f} points")

    comparison_fig_path = os.path.join(OUT_DIR, "baseline_vs_final_classification.png")
    save_final_comparison_figure(x_te, y_te, baseline_pred, final_pred, comparison_fig_path)

    summary = {
        "run_id": RUN_ID,
        "out_dir": OUT_DIR,
        "data_split_summary": split_summary,
        "baseline_val_accuracy": val_acc,
        "baseline_test_accuracy": float(test_acc),
        "ga_best_val_accuracy": float(best_ga_val),
        "final_test_accuracy": float(final_acc),
        "improvement_points": float((final_acc - test_acc) * 100.0),
        "selected_candidate_label": winner["label"],
        "selected_candidate_val_acc_mean": float(winner["val_acc_mean"]),
        "selected_candidate_val_acc_std": float(winner["val_acc_std"]),
        "selected_candidate_val_acc_runs": winner["val_acc_runs"],
        "selected_params": winner["params"],
        "comparison_figure_path": comparison_fig_path,
        "baseline_learning_curves_path": baseline_curves_path,
        "final_learning_curves_path": final_curves_path,
    }
    summary_path = os.path.join(OUT_DIR, "summary.json")
    with open(summary_path, "w", encoding="utf-8") as f:
        json.dump(summary, f, indent=2)

    print("\nArtifacts:")
    print(f"- CSVs per generation: {OUT_DIR}/gen_XX_population.csv")
    print(f"- Transition heatmaps: {FIG_DIR}/gen_XX_transition.png")
    print(f"- Dataset split overview: {split_fig_path}")
    print(f"- Baseline learning curves: {baseline_curves_path}")
    print(f"- Final model learning curves: {final_curves_path}")
    print(f"- Final comparison figure: {comparison_fig_path}")
    print(f"- Summary metrics: {summary_path}")
    print(f"- Run directory: {OUT_DIR}")
    display_saved_figure(split_fig_path, "Displayed dataset split overview")
    display_saved_figure(baseline_curves_path, "Displayed baseline learning curves")
    display_saved_figure(final_curves_path, "Displayed final model learning curves")
    display_saved_figure(comparison_fig_path, "Displayed final comparison figure")
    print("\nDone.")

if __name__ == "__main__":
    main()
```
